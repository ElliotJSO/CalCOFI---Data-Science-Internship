```python
#Cluster Model by Kaylin Roberts

from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans

#creating a separate table with only temperature and depth variables 
depth_temp = data_table.select("Temperature_degC","Depth(m)")

#cleaning up table to have data only where depth is <= 500 && temperature is >= 0
depth_temp_table = depth_temp.where("Depth(m)",are.below_or_equal_to(500)).where("Temperature_degC",are.above_or_equal_to(0))

#create elbow graph to find best number of clusters for model 

temp_array = depth_temp_table.column(0).reshape(-1,1)
depth_array = depth_temp_table.column(1)

k_range = range(1,10)
sse = []
for k in k_range:
   km = KMeans(n_clusters=k)
   km.fit(temp_array,depth_array)
   #finds the sum of squared error for the fitted graph 
   sse.append(km.inertia_)

#look at "elbow" of the graph to detemine best num of clusters
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.plot(k_range,sse)



```

```python
#create cluster model with n_clusters = 3
km = KMeans(n_clusters=3)

#scaling the data so model is more precise (changing depth and temp to be in range 0-1)
scaler = MinMaxScaler()
scaler.fit(temp_array)
scaler.fit(depth_array.reshape(-1,1))
scaled_temp = scaler.transform(temp_array)
scaled_depth = scaler.transform(depth_array.reshape(-1,1))
depth_temp_table = depth_temp_table.with_columns('Scaled Temperature',scaled_temp,'Scaled Depth', scaled_depth)

#adding scaled depth and temp to table 
depth_array2 = depth_temp_table.column(3)
temp_array2 = depth_temp_table.column(2).reshape(-1,1)
km2 = KMeans(n_clusters=3)
y_predicted2 = km2.fit_predict(depth_array2,temp_array2)
depth_temp_table = depth_temp_table.with_column('Clusters After Scale', y_predicted2)

#graphing the model
#seperating the data by cluster
df1_2 = depth_temp_table.where(4,are.equal_to(0))
df2_2 = depth_temp_table.where(4,are.equal_to(1))
df3_2 = depth_temp_table.where(4,are.equal_to(2))

#plotting each cluster with different colors
plt.scatter(df1_2.column(2),df1_2.column(3),color = 'green')
plt.scatter(df2_2.column(2),df2_2.column(3),color = 'blue')
plt.scatter(df3_2.column(2),df3_2.column(3),color = 'black')

plt.xlabel('Temperature')
plt.ylabel('Depth')
```
